```python
# Hough Transform : Lines

# Image Space (Coordinate system in x and y) : y = m0 x + c0
# Hough Space (Coordinate system in m and c) : c = -x0 m + y0 
# m0 & c0 describe the line in x-y plane and make a point in m-c plane
# Similarly, x0 & y0 describe the line in m-c plane and make a point in x-y plane

# Consider some points in Image Space, through which we must pass a line.
# In Hough Space, these are lines.
# Visualize the Hough Space as if it is on graph paper.
# Select the bin with the majority of interecting lines (voting).
# Note: An 2D array kept to count votes is call a Hough accumulator array.
# Take the center point of this bin.
# This point represents the desired line in Image Space.

# If Image Space is represented with polar coordinates, then p = x.cos(t) - y.sin(t)
# So a sinusoid in Image Space is a point in Hough Space
# Similarly, a point in Image Space is a sinusoid in Hough Space
# In this case, the algorithm is outlined below

from skimage import io
img = io.imread("hough_line_algorithm.png")
io.imshow(img)
```
